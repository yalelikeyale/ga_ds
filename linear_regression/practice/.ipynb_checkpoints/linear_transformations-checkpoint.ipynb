```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def func(x, a, b, c):
    return - a * np.exp(-b * x) + c

x = np.linspace(0, 4, 50)
y = func(x, 2.5, 1.3, 0.5)

plt.scatter(x, y, alpha=0.5);
```

```python
def func(x, a, b, c):
    return a * np.square(x) + c

x = np.linspace(0, 4, 50)
y = func(x, 2.5, 1.3, 0.5)

plt.scatter(x, y, alpha=0.5);
```
